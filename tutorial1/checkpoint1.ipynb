<h2> Feature Engineering </h2>

After exploring the data, you can choose to modify features that you would use to predict the performance of the students on your chosen response variable.

```python
data["INSTRUCTION_EXP_PER_REVENUE"] = data["INSTRUCTION_EXPENDITURE"] / data["ENROLL"]
data["PERCENT_OF_REVENUE"] = data["INSTRUCTION_EXPENDITURE"] / data["TOTAL_REVENUE"]
data.dropna()
```

<h2> Data Creation </h2>

_Use this space to create train/test data_

```python
from sklearn.model_selection import train_test_split
```

```python
X = data[["AVG_MATH_8_SCORE", "AVG_MATH_4_SCORE", "AVG_READING_8_SCORE"]].dropna()
y = data.loc[X.index]["AVG_READING_4_SCORE"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
X, y, test_size=3, random_state=42)
y.fillna(y.median(), inplace=True)
```

<h2> Prediction </h2>

ML Models [Resource](https://medium.com/@vijaya.beeravalli/comparison-of-machine-learning-classification-models-for-credit-card-default-data-c3cf805c9a5a)

```python
# import your sklearn class here
from sklearn.linear_model import LinearRegression
```

```python
# create your model here
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

## Evaluation

Choose some metrics to evaluate the performance of your model, some of them are mentioned in the tutorial.

```python
model.score(X_test, y_test)
```

We have copied over the graphs that visualize the model's performance on the training and testing set. 

Change `col_name` and modify the call to `plt.ylabel()` to isolate how a single predictor affects the model.

```python
col_name = 'AVG_MATH_4_SCORE'

f = mlb.figure(figsize=(12,6))
mlb.scatter(X_train[col_name], y_train, color = "red")
mlb.scatter(X_train[col_name], model.predict(X_train), color = "green")

mlb.legend(['True Training','Predicted Training'])
mlb.xlabel(col_name)
mlb.ylabel('Reading 4 Score')
mlb.title("Model Behavior On Training Set")
```

```python
col_name = 'AVG_MATH_4_SCORE'

f = mlb.figure(figsize=(12,6))
mlb.scatter(X_test[col_name], y_test, color = "blue")
mlb.scatter(X_test[col_name], model.predict(X_test), color = "black")

mlb.legend(['True testing','Predicted testing!'])
mlb.xlabel(col_name)
mlb.ylabel('Reading 4 Score')
mlb.title("Model Behavior On Training Set")
```
